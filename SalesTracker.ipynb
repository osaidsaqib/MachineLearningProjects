```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'sales_data.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head(10))
```

```python
totalSales = df.groupby('Product')['Quantity'].sum()
print(totalSales)
```

```python
df['Revenue'] = df['Quantity']*df['Price']
totalRevenue = df.groupby('Product')['Revenue'].sum()
print(totalRevenue)
```

```python
import matplotlib.pyplot as plt
years = [1980, 1990, 2000, 2010, 2020]
power_UK = [120, 250, 400, 550, 900]
power_USA = [15, 35, 70, 120, 310]
plt.plot(years, power_UK, color='blue', marker='x', markersize=8, label='UK')
plt.plot(years, power_USA, color='red', linestyle='dotted', linewidth=2, label='USA')
plt.xlabel('Years')
plt.ylabel('Power Consumption (kWh)')
plt.title('Power Consumption of UK and USA')
plt.legend()
plt.show()
```
