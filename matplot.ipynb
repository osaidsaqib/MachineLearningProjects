```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
quarter = ['Q1', 'Q2', 'Q3', 'Q4']
sales_data_2020 = {'A':[10, 23, 30, 40], 'B':[15,25,45,50]}
sales_data_2021 = {'A':[12, 24,40,52], 'B':[3,17, 50, 40]}
```

```python
plt.plot(quarter, sales_data_2020['A'], label='Product A sales 2020', marker = 'o')
plt.plot(quarter, sales_data_2020['B'], label='Product B sales 2020', marker = 's')
plt.plot(quarter, sales_data_2021['A'], label='Product A sales 2021', marker = '^')
plt.plot(quarter, sales_data_2021['B'], label='Product B sales 2021', marker = 's')
plt.xlabel("Quarter")
plt.ylabel("Sales in $ 2020")
plt.title("Sales data for product A and B")
plt.legend()
plt.show()
```

```python
#city_years = ['2020', '2021']
#city_data = {'city A':[1.5, 1.7], 'city B':[2.1, 2.4], 'city C':[1.8, 1.9], 'city D':[2.0, 2.2], 'city E':[1.6, 1.7]}
cities = ['city a', 'city b', 'city c', 'city d', 'city e']
population_2020 = [1.5, 2.1, 1.8, 2.0, 1.6]
population_2021 = [1.7, 2.4, 1.9, 2.2, 1.7]
```

```python
barwidth = 0.35
index = np.arange(len(cities))

```

```python
plt.bar(index, population_2020, barwidth, label = '2020')
plt.bar(index + barwidth, population_2021, barwidth, label = '2021')
plt.xlabel('Cities')
plt.ylabel('Population (in millions)')
plt.title('Population comparision between 2020-2021')
```

```python
label = ['North', 'South', 'East', 'West']
size = [35, 25, 20, 20]
explode = (0.1, 0, 0, 0)
```

```python
plt.pie(size, explode = explode, labels = label, autopct = '%1.1f%%', shadow = True, startangle = 140)
plt.axis('equal')
plt.title('Market share over geographical area')
plt.show
```
