```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'mysales.csv'
df = pd.read_csv(file_path, index_col = 'Item')
```

```python
print(df)
```

```python
df.plot(kind = 'bar', color = 'red')
plt.xlabel('Item')
plt.ylabel('sales')
plt.title('Total rev per item')
plt.xticks(rotation = 50)
plt.tight_layout()
plt.show()
```
