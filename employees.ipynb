```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'Emplyees.csv'
df = pd.read_csv(file_path)
```

```python
print(df)
```

```python
it_department = df['Department'].value_counts()
```

```python
print(it_department)
```

```python
avg_salary = df.groupby('Department')['Salary'].mean()
print(avg_salary)
```

```python
df['Date of joining'] = pd.to_datetime(df['Date of joining'])
after2020 = df[df['Date of joining']> '1/1/2020']
```

```python
print(after2020)
```

```python
orderofdate = df.sort_values(by='Date of joining')
print(orderofdate)
```

```python
plt.pie(it_department, labels=['Sales', 'It', 'Marketing', 'HR', 'Reaserch and Development'])
plt.axis('equal')
plt.show()
```
