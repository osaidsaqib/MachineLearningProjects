```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'sales_data.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head(10))
```

```python
totalSales = df.groupby('Product')['Quantity'].sum()
print(totalSales)
```

```python
df['Revenue'] = df['Quantity']*df['Price']
totalRevenue = df.groupby('Product')['Revenue'].sum()
print(totalRevenue)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
janSales = df[df['Date'].dt.month == 1]
print(janSales)
```

```python
totalRevenue.plot(kind = 'bar', color = 'red')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Total rev per item')
plt.xticks(rotation = 50)
plt.tight_layout()
plt.show()
```
